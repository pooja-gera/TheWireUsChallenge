```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df=pd.read_csv('input_covid.csv')
```

```python
df.head()
```

```python
a = df['chou_fasman'][0:50]
b = df['kolaskar_tongaonkar'][0:50]
c = df['emini'][0:50]
d = df['parker'][0:50]
e = df['isoelectric_point'][0:50]
f = df['aromaticity'][0:50]
g = df['hydrophobicity'][0:50]

#Giving title to the chart
plt.title("Properties of 6VYB_A proteins with different peptide sequences")

#Giving axis labels
plt.xlabel('Data number') 
plt.ylabel('Value') 

#Plotting various features
plt.plot(a,label="chou fasman")
plt.plot(b,label="kolaskar tongaonkar")
plt.plot(c,label="emini")
plt.plot(d,label="parker")
plt.plot(e,label="isoelectric point")
plt.plot(f,label="aromaticity")
plt.plot(g,label="hydrophobicity")

#Adding legend using the labels of each axis
plt.legend()

#Show grid
plt.grid()
```
