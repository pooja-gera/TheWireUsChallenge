```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

from time import sleep
from random import randint

# to select only english
headers = {"Accept-Language": "en-US,en;q=0.5"}

# data storage
titles = []
years = []
time = []
imdb_ratings = []
metascores = []
votes = []
us_gross = []

# define the number of pages you want
pages = np.arange(1, 11)

for page in pages:
    # make GET requests
    page = requests.get("https://www.imdb.com/search/title/?groups=top_1000&start=" +
                        str(page) + "&ref_=adv_nxt", headers=headers)

    soup = BeautifulSoup(page.text, 'html.parser')
    movie_div = soup.find_all('div', class_='lister-item mode-advanced')

    # to give time to load page
    sleep(randint(2, 10))

    for item in movie_div:

        name = item.h3.a.text
        titles.append(name)

        year = item.h3.find('span', class_='lister-item-year').text
        years.append(year)

        runtime = item.p.find('span', class_='runtime') if item.p.find(
            'span', class_='runtime') else ''
        time.append(runtime)

        imdb = float(item.strong.text)
        imdb_ratings.append(imdb)

        m_score = item.find('span', class_='metascore').text if item.find(
            'span', class_='metascore') else ''
        metascores.append(m_score)

        nv = item.find_all('span', attrs={'name': 'nv'})

        vote = nv[0].text
        votes.append(vote)

        grosses = nv[1].text if len(nv) > 1 else ''
        us_gross.append(grosses)

movies = pd.DataFrame({
    'movie': titles,
    'year': years,
    'imdb': imdb_ratings,
    'metascore': metascores,
    'votes': votes,
    'us_gross(M)': us_gross,
    'time(Min.)': time
})
```
