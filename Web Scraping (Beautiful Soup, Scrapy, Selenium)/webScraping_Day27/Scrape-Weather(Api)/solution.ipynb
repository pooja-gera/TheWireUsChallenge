```python
# load necessary libraries
import requests
import json
import pandas as pd
```

```python
# loade the dataset provided
df=pd.read_csv("/Users/jappanjeetsingh/Downloads/datasets/IndianCities.csv")
df["city"]
```

```python
# define data containers
description=[]
temp=[]
feels_like=[]
pressure=[]
humidity=[]
visibility=[]
wind_speed=[]
# put your api key
api_key=""
for city in df["city"]:
    url=f"http://api.openweathermap.org/data/2.5/weather?q={city},IN&appid={api_key}"
    response=requests.get(url)
    data_json=json.loads(response.content)
#     save the succesfully received data (city found)
    if list(data_json.values())[1]!="city not found":
        description.append(json.dumps(data_json["weather"][0]["description"]))
        temp.append(float(json.dumps(data_json["main"]["temp"]))-273.00)
        feels_like.append(float(json.dumps(data_json["main"]["feels_like"]))-273.00)
        pressure.append(json.dumps(data_json["main"]["pressure"]))
        humidity.append(json.dumps(data_json["main"]["humidity"]))
        visibility.append(json.dumps(data_json["visibility"]))
        wind_speed.append(json.dumps(data_json["wind"]["speed"]))
```

```python
# add data to the dataframe (df)
df["description"]=pd.Series(description)
df["temperature(°C)"]=pd.Series(temp)
df["feels-like(°C)"]=pd.Series(feels_like)
df["pressure"]=pd.Series(pressure)
df["humidity"]=pd.Series(humidity)
df["visibility"]=pd.Series(visibility)
df["wind-speed"]=pd.Series(wind_speed)
```
