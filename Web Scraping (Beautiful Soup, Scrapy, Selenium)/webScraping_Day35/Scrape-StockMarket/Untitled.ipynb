```python
from bs4 import BeautifulSoup
import json
import requests
import pandas as pd
```

```python
start="2017-01-01"
end="2018-01-01"
url=f"https://www.quandl.com/api/v3/datasets/WIKI/AAPL.json?start_date={start}&end_date={end}"
```

```python
response=requests.get(url)
response # should be a 200 - good response
```

```python
data=json.loads(response.content)
data["dataset"]["column_names"]
data["dataset"]["data"]
# create the dataframe
df = pd.DataFrame(data['dataset']['data'], columns = data['dataset']['column_names'])
```
