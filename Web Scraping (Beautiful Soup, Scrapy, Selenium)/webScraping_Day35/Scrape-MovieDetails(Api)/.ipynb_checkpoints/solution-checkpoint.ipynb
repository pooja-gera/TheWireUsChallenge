```python
import requests
import pandas as pd
import json
```

```python
df=pd.read_csv("./movies.csv")
```

```python
# cleaning the enwanted characters in the movie titles
df[df.columns] = df.apply(lambda x: x.str.strip())
```

```python
# add your api key
API_KEY="c1f48d66"
# storage containers
Year=[]
Rated=[]
Released=[]
Runtime=[]
Genre=[]
Director=[]
Actors=[]
Language=[]
Country=[]
Awards=[]
imdbRating=[]
imdbVotes=[]
BoxOffice=[]
for title in df["movie_title"]:
    url=f"http://www.omdbapi.com/?t={title}&apikey={str(API_KEY)}"
    response=requests.get(url)
    json_obj=json.loads(response.content)
    if list(json_obj.values())[1]!='Movie not found!':
        print("enter")
        Year.append(json.dumps(json_obj["Year"]))
        Rated.append(json.dumps(json_obj["Rated"]))
        Released.append(json.dumps(json_obj["Released"]))
        Runtime.append(json.dumps(json_obj["Runtime"]))
        Genre.append(json.dumps(json_obj["Genre"]))
        Director.append(json.dumps(json_obj["Director"]))
        Actors.append(json.dumps(json_obj["Actors"]))
        Language.append(json.dumps(json_obj["Language"]))
        Country.append(json.dumps(json_obj["Country"]))
        Awards.append(json.dumps(json_obj["Awards"]))
        imdbRating.append(json.dumps(json_obj["imdbRating"]))
        imdbVotes.append(json.dumps(json_obj["imdbVotes"]))
#         BoxOffice.append(json.dumps(json_obj["BoxOffice"]))
```

```python
# storing data into the data frame by first converting the lists to pandas series to handle NaN values
df["Year"]=pd.Series(Year)
df["Rated"]=pd.Series(Rated)
df["Released"]=pd.Series(Released)
df["Runtime"]=pd.Series(Runtime)
df["Genre"]=pd.Series(Genre)
df["Director"]=pd.Series(Director)
df["Actors"]=pd.Series(Actors)
df["Language"]=pd.Series(Language)
df["Country"]=pd.Series(Country)
df["Awards"]=pd.Series(Awards)
df["imdbRating"]=pd.Series(imdbRating)
df["imdbVotes"]=pd.Series(imdbVotes)
```
